```python
import pandas as pd
#install and import the jellyfish library
!pip install jellyfish
import jellyfish


!wget https://norvig.com/ngrams/count_1w.txt #unigram corpus
filePath1 = "/content/count_1w.txt"
unigrams_df = pd.read_csv(filePath1,sep='\t',header=None, names=['unigram','count']) #keep_default_na to be set here to avoid NaNs
print(f'Number of unigrams: {unigrams_df.shape[0]}')
display(unigrams_df.head(100))
```

Task 1.B

```python
def nonwordspellingcorrection(nonword):
  candidate=[]
  total=unigrams_df['count'].sum()
  for word in unigrams_df['unigram']:
    if isinstance(word, float):
            continue
    d=jellyfish.damerau_levenshtein_distance(nonword,word)
    if d <= 1:
      count = int(unigrams_df.loc[unigrams_df['unigram'] == word, 'count'])
      prob = count / total
      candidate.append((word, count, prob))
  print("Non word:", nonword)
  print("word \t Frequency\tP(word)\n")
  for w, c, p in candidate:
        print(f"{w}   {c}   {p:.8f}")
  return
nonwordspellingcorrection('acress')
```

```python
from textblob import TextBlob
def textb(word):
  b=TextBlob(word)
  print("string =", word)
  s = b.words[0].spellcheck()
  print(s)
  return
textb('acress')
```

```python
def correct(sentence):
    b = TextBlob(sentence)
    corrected = b.correct()
    return corrected

sen="I have good speling"
crrct_sentence=correct(sen)
print("sentence =", sen)
print("corrected sentence =", crrct_sentence)
```
