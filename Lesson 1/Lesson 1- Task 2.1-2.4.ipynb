```python
from tensorflow.keras.preprocessing.text import Tokenizer
!wget http://www.gutenberg.org/files/100/100-0.txt
filePath = "/content/100-0.txt"
ShakespeareFile = open(filePath, "r")
ShakespeareContent = ShakespeareFile.read()
ShakespeareContent = ShakespeareContent.splitlines() # The splitlines() method splits a string into a list. The splitting is done at line breaks.
print(f'list of first 50 lines: {ShakespeareContent[0:50]}')
tk=Tokenizer()
tk.fit_on_texts(ShakespeareContent)
word_index=tk.word_index
count=tk.word_counts
print(f'Found {(word_index)} unique tokens.')
print("Total number of tokens = ",sum(count.values()))
print("Total number of types = ",len(word_index))
c=0
for k,v in word_index.items():
  print(k, "is ranked",c+1,"with a frequency of ",count[k])
  c=c+1
  if c==10:
    break
```

Task 2.3

```python
from nltk.stem import PorterStemmer
import nltk
nltk.download('wordnet')
from nltk.stem.wordnet import WordNetLemmatizer
ps = PorterStemmer()
lemmatizer = WordNetLemmatizer()
l1=[]
for k,v in word_index.items():
  l1.append(ps.stem(k))
print(l1)
l2=[]
for k,v in word_index.items():
  l2.append(lemmatizer.lemmatize(k))
print(l2)
print('\n\nTotal number of Types after stemming : ',len(set(l1)))
print('\nTotal number of Types after Lemmatization : ',len(set(l2)))
```

validity of this arithmetic expression:

total_number_of_types > total_number_of_lemmatized_types > total_number_of_stemmed_types

```python
if len(word_index)>len(set(l2))>len(set(l1)):
  print("The validity of the provided Arithmetic expression is True\n")
  print("total_number_of_types > total_number_of_lemmatized_types > total_number_of_stemmed_types")
else:
    print("The expected validity not obtained")
```
