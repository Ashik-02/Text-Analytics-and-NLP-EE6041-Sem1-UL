```python
import time

#How many insertions, deletions, or substitutions does it take to turn x into y?
def edDistRecursive(x, y):
    # If either x or y is empty, then one or more (depending on their length) INSERTIONS or DELETIONS are needed to transform x to y.
    if len(x) == 0:
        #print(f'\t Converting "{x}" TO "{y}" requires {len(y)} INSERTIONS, therefor it costs {len(y)}')
        return len(y) * 2
    if len(y) == 0:
        #print(f'\t Converting "{x}" TO "{y}" requires {len(x)} DELETIONS, therefor it costs {len(x)}')
        return len(x) * 2

    # When neither x nor y is empty, we must compute the cost of each operation and find the least costly one.
    if x[-1] == y[-1]:
        delta = 0
    else:
        if (x[-1].isalpha() and y[-1].isalpha()) or (x[-1].isdigit() and y[-1].isdigit()):
            delta = 2
        else:
            delta = 4

    diagonal_or_substitution_cost = edDistRecursive(x[:-1], y[:-1]) + delta  #what's the cost of SUBSTITUTING the last character of x with the last character of y
    vertical_or_deletion_cost     = edDistRecursive(x[:-1], y)      + 2      #what's the cost of DELETING the last character of x
    horizontal_or_insertion_cost  = edDistRecursive(x, y[:-1])      + 2      #what's the cost of INSERTING the last character of x into y# substitution

    # What's the least costly operation?
    return min(diagonal_or_substitution_cost, vertical_or_deletion_cost, horizontal_or_insertion_cost)


start = time.time()
value = edDistRecursive("a2c", "ab1")
end = time.time()

print('Weighted Edit Distance between "a2c" and "ab1":', value)
print("total time:", end - start, "sec")
```
